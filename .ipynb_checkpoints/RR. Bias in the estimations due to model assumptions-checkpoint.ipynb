## Preliminaries

```python
placettLuceDirichlet = PlackettLuceDirichlet(alpha=[1], num_samples=1000, seed=seed)
mallowsModel = MallowsModel(num_samples=1000, seed=seed)
```

```python
def calculate_top_ranking_probs(orderings, num_samples=10000):
  num_instances, num_algorithms = orderings.shape

  # Priors definition
  dirichlet_alpha_pl = num_algorithms * [1]
    
  # Update model priors and other hyper-parameters
  placettLuceDirichlet.alpha = dirichlet_alpha_pl
  placettLuceDirichlet.num_samples = num_samples
  mallowsModel.num_samples = num_samples
  
    
  pld = placettLuceDirichlet.calculate_top_ranking_probs(orderings)
  mm  = mallowsModel.calculate_top_ranking_probs(orderings)

  return pld, mm
```

```python
def samplePL(theta):
    pi = []
    n = len(theta)
    pool = [i for i in range(n)]

    while len(pi) < n:
        probs = [theta[i] / np.sum([theta[j] for j in pool]) for i in pool]
        item = np.random.choice(pool, p=probs)
        pi.append(item)
        pool.remove(item)

    return pi
```

```python
def sampleMM(center, phi):
    num_items = len(center)
    rng = np.arange(0, num_items - 1)
    psi = (1 - phi ** (num_items - rng)) / (1 - phi)
    vprobs = np.zeros((num_items, num_items))

    for i in rng:
      vprobs[i][0] = 1.0 / psi[i]
      for j in range(1, num_items - i):
        vprobs[i][j] = phi ** j / psi[i]

    inv = [np.random.choice(num_items, p=vprobs[i,:]) for i in rng] + [0] 
    rem = list(range(num_items))
    perm = np.full(num_items, np.nan, dtype=np.int32)

    for i in range(num_items):
      perm[i] = rem[inv[i]]
      rem.pop(inv[i])

    perm = perm[center]
    return perm
```

## Synthetic Data Analysis

```python
num_samples = 10
theta = [4, 9, 1, 5]
permus = np.array([samplePL(theta) for _ in range(num_samples)]) + 1
orderings = np.argsort(permus) + 1
probs = calculate_top_ranking_probs(permus)
p_top_ranking = theta / np.sum(theta)
model_names = ["PLD", "MM"]
algorithm_names=['$A_1$', '$A_2$', '$A_3$', '$A_4$']
```

```python
fig = plt.figure()
fig, axs = plt.subplots(1, 4, figsize=(15, 3))

plotter = Plot()
plotter.plot_top_ranking_probs(model_names, algorithm_names, probs, p_top_ranking, axs)

plt.tight_layout()

fig.savefig("img/RR.Bias.Estimation.10.pdf", bbox_inches='tight')
```
