## Preliminaries

```python
bradleyTerry = BradleyTerry([1], num_samples=1000, seed=seed)
placettLuceDirichlet = PlackettLuceDirichlet(alpha=[1], num_samples=1000, seed=seed)
placettLuceGamma = PlackettLuceGamma(alpha=0.5, beta=0.5, num_samples=1000, seed=seed)
```

```python
def get_posterior_uninformed_prior(orderings, weights, scores, num_samples=1000):
  num_instances, num_algorithms = orderings.shape

  # Priors definition
  dirichlet_alpha_bt = num_algorithms * [1]
  dirichlet_alpha_pl = num_algorithms * [1]
  gamma_alpha_pl = 0.5
  gamma_beta_pl = 0.5
    
  # Update model priors and other hyper-parameters
  bradleyTerry.alpha = dirichlet_alpha_bt
  bradleyTerry.num_samples = num_samples
  placettLuceDirichlet.alpha = dirichlet_alpha_pl
  placettLuceDirichlet.num_samples = num_samples
  placettLuceGamma.alpa = gamma_alpha_pl
  placettLuceGamma.beta = gamma_beta_pl
  placettLuceGamma.num_samples = num_samples
  
  pld_top = placettLuceDirichlet.calculate_top_ranking_probs(orderings, weights)
  plg_top = placettLuceGamma.calculate_top_ranking_probs(orderings, weights)
  bt_top  = bradleyTerry.calculate_top_ranking_probs(orderings, weights)
    
  pld_better = placettLuceDirichlet.calculate_better_than_probs(orderings, weights)
  plg_better = placettLuceGamma.calculate_better_than_probs(orderings, weights)
  bt_better  = bradleyTerry.calculate_better_than_probs(orderings, weights)

  pld_topk = placettLuceDirichlet.calculate_top_k_probs(orderings, weights)
  plg_topk = placettLuceGamma.calculate_top_k_probs(orderings, weights)
  bt_topk  = bradleyTerry.calculate_top_k_probs(orderings, weights)
  
  return bt_top, pld_top, plg_top, bt_better, pld_better, plg_better, bt_topk, pld_topk, plg_topk
```

```python
def get_posterior_deceptive_prior(orderings, weights, scores, num_samples=1000):
  num_instances, num_algorithms = orderings.shape

  alpha = np.argsort(np.sum(scores, axis=0))
  alpha = alpha + 1

  # Priors definition
  dirichlet_alpha_bt = alpha
  dirichlet_alpha_pl = alpha
  gamma_alpha_pl = 0.5
  gamma_beta_pl = 0.5
    
  # Update model priors and other hyper-parameters
  bradleyTerry.alpha = dirichlet_alpha_bt
  bradleyTerry.num_samples = num_samples
  placettLuceDirichlet.alpha = dirichlet_alpha_pl
  placettLuceDirichlet.num_samples = num_samples
  placettLuceGamma.alpa = gamma_alpha_pl
  placettLuceGamma.beta = gamma_beta_pl
  placettLuceGamma.num_samples = num_samples
  
  pld_top = placettLuceDirichlet.calculate_top_ranking_probs(orderings, weights)
  plg_top = placettLuceGamma.calculate_top_ranking_probs(orderings, weights)
  bt_top  = bradleyTerry.calculate_top_ranking_probs(orderings, weights)
    
  pld_better = placettLuceDirichlet.calculate_better_than_probs(orderings, weights)
  plg_better = placettLuceGamma.calculate_better_than_probs(orderings, weights)
  bt_better  = bradleyTerry.calculate_better_than_probs(orderings, weights)

  pld_topk = placettLuceDirichlet.calculate_top_k_probs(orderings, weights)
  plg_topk = placettLuceGamma.calculate_top_k_probs(orderings, weights)
  bt_topk  = bradleyTerry.calculate_top_k_probs(orderings, weights)
  
  return bt_top, pld_top, plg_top, bt_better, pld_better, plg_better, bt_topk, pld_topk, plg_topk
```

```python
def get_posterior_informed_prior(orderings, weights, scores, num_samples=1000):
  num_instances, num_algorithms = orderings.shape

  alpha = np.argsort(np.argsort(np.sum(scores, axis=0)))
  alpha = alpha + 1
    

  # Priors definition
  dirichlet_alpha_bt = alpha
  dirichlet_alpha_pl = alpha
  gamma_alpha_pl = 0.5
  gamma_beta_pl = 0.5
    
  # Update model priors and other hyper-parameters
  bradleyTerry.alpha = dirichlet_alpha_bt
  bradleyTerry.num_samples = num_samples
  placettLuceDirichlet.alpha = dirichlet_alpha_pl
  placettLuceDirichlet.num_samples = num_samples
  placettLuceGamma.alpa = gamma_alpha_pl
  placettLuceGamma.beta = gamma_beta_pl
  placettLuceGamma.num_samples = num_samples
  
  pld_top = placettLuceDirichlet.calculate_top_ranking_probs(orderings, weights)
  plg_top = placettLuceGamma.calculate_top_ranking_probs(orderings, weights)
  bt_top  = bradleyTerry.calculate_top_ranking_probs(orderings, weights)
    
  pld_better = placettLuceDirichlet.calculate_better_than_probs(orderings, weights)
  plg_better = placettLuceGamma.calculate_better_than_probs(orderings, weights)
  bt_better  = bradleyTerry.calculate_better_than_probs(orderings, weights)

  pld_topk = placettLuceDirichlet.calculate_top_k_probs(orderings, weights)
  plg_topk = placettLuceGamma.calculate_top_k_probs(orderings, weights)
  bt_topk  = bradleyTerry.calculate_top_k_probs(orderings, weights)
  
  return bt_top, pld_top, plg_top, bt_better, pld_better, plg_better, bt_topk, pld_topk, plg_topk
```

```python
# File names of the data for the different algorithms
algorithms = ['A', 'B', 'AGA', 'VNS', 'NVNS']

# Algorith names
algorithm_names = ['GM-EDA', 'HGM-EDA', 'AGA', 'VNS', 'NVNS']

# Load the data from file
orderingsT, scoresT = load_permus_from_CEB('data/CEB19/T', algorithms)
orderingsR, scoresR = load_permus_from_CEB('data/CEB19/R', algorithms)

# Concatenate Taillard instances with the Random instances
scores = np.concatenate((scoresT, scoresR), axis=0)
orderings = np.concatenate((orderingsT, orderingsR), axis=0)

sam_orderings, sam_scores = get_sample(orderings, scores, 100)
counts = get_counts(sam_scores)
sam_orderings, sam_weights = unroll(sam_orderings, counts)
```

```python
# Calculate the empirical probabilities from the data
p_top_ranking = empirical_top_ranking_probs(scores)
p_better_than = empirical_better_than(scores)
```

```python
uninformed = get_posterior_uninformed_prior(sam_orderings, sam_weights, scores)
informed = get_posterior_informed_prior(sam_orderings, sam_weights, scores)
deceptive = get_posterior_deceptive_prior(sam_orderings, sam_weights, scores)
```
